```python
from qfa.domain import FeedbackRecordModel
from qfa.adapters.presidio_anonymizer import PresidioAnonymizer
from qfa.domain.ports import handle_anonymization


class MyOrchestrator:
    def __init__(self, anonymizer: PresidioAnonymizer):
        self.anonymizer = anonymizer

    @handle_anonymization
    def some_method(self, request: FeedbackRecordModel) -> FeedbackRecordModel:
        print(request)
        return request


example_record = FeedbackRecordModel(
    id="hello",
    content="This is a sample feedback containing sensitive information. My name is Cas Teeuwen and I live in Amsterdam.",
)

anonymizer = PresidioAnonymizer()
orchestrator = MyOrchestrator(anonymizer)
anonymized_record = orchestrator.some_method(example_record)
```
