```python
from qfa.adapters.presidio_anonymizer import PresidioAnonymizer

anonymizer = PresidioAnonymizer()
```

```python
def anon_and_deanon(input: str) -> None:
    print("-----------------------")
    anonymized, mapping = anonymizer.anonymize(input)
    print(anonymized)
    print(anonymizer.deanonymize(anonymized, mapping))


anon_and_deanon("Hello, my name is John and I'm from Amsterdam")
anon_and_deanon("Привіт, мене звати Джон, і я з Амстердама.")
anon_and_deanon("Привет, меня зовут Джон, и я из Амстердама.")
anon_and_deanon("مرحبًا، اسمي جون وأنا من أمستردام.")
anon_and_deanon("Bonjour, je m'appelle John et je viens d'Amsterdam.")
anon_and_deanon("Hola, me llamo John y soy de Ámsterdam.")
```
