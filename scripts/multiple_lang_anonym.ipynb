```python
from langdetect import detect

LANGUAGES_AND_ANONYMIZATION_MODEL_PAIRINGS = [
    {"lang_code": "en", "model_name": "en_core_web_sm"},
    {"lang_code": "fr", "model_name": "fr_core_news_sm"},
    {"lang_code": "uk", "model_name": "uk_core_news_sm"},
    {"lang_code": "ru", "model_name": "ru_core_news_sm"},
    {"lang_code": "xx", "model_name": "xx_ent_wiki_sm"},
]


def detect_language(text: str) -> str:
    language_shortcode = detect(text)
    if language_shortcode not in [
        ent["lang_code"] for ent in LANGUAGES_AND_ANONYMIZATION_MODEL_PAIRINGS
    ]:
        return "xx"
    return language_shortcode


def anonymize_with_lang_detection(input_text: str):
    detected_language = detect_language(input_text)
    print(f"{input_text=} === {detected_language=}")
    nlp_engine = NlpEngineProvider(
        nlp_configuration={
            "nlp_engine_name": "spacy",
            "models": LANGUAGES_AND_ANONYMIZATION_MODEL_PAIRINGS,
        }
    ).create_engine()
    analyzer_engine = AnalyzerEngine(nlp_engine=nlp_engine)
    result = analyzer_engine.analyze(input_text, language=detected_language)
    print(f"{result=}")


anonymize_with_lang_detection("Hello my name is John")
anonymize_with_lang_detection("Bonjour mon nom est John")
anonymize_with_lang_detection("Привіт, мене звати Джон")
anonymize_with_lang_detection("Привет, меня зовут Джон")
anonymize_with_lang_detection("مرحبًا، اسمي جون")
```
